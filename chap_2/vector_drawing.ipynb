```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def draw_origin():
    plt.scatter(0,0, marker='x', color='k')
    
def draw_axes(ax):
    ax.axhline(lw=1, color='k')
    ax.axvline(lw=2, color='k')
    
def draw_grid(ax):
    ax.set_xticks(np.arange(-20,20, 1))
    ax.set_yticks(np.arange(-20, 20, 1))
    plt.grid(True)
    ax.set_axisbelow(True)
    
def set_bounds(xmin, xmax, ymin, ymax):
    plt.xlim((xmin,xmax))
    plt.ylim((ymin,ymax))
    
def draw_points(dino_vectors):
    xs = list(map(lambda x: x[0], dino_vectors))
    ys = list(map(lambda x: x[1], dino_vectors))
    plt.scatter(xs,ys)
    
def display():
    plt.show()
    
def draw_segment(p1,p2):
    x1, y1 = p1
    x2, y2 = p2
    plt.plot([x1,x2], [y1, y2], color='g')
    
```

```python
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]
```

```python
fig, ax = plt.subplots()
draw_axes(ax)
draw_grid(ax)
draw_origin()
set_bounds(-8, 8, -8, 8)
draw_points(dino_vectors)
display()
```

```python
fig, ax = plt.subplots()
draw_axes(ax)
draw_grid(ax)
draw_origin()
set_bounds(-8, 8, -8, 8)
draw_points(dino_vectors)
draw_segment((6,4), (3,1))
display()
```

```python
'''
EXERCISE
Build another abstraction function connect_the_dots(plane, vectors) that takes the target plan and 
a list of vectors and draws segments connecting each vector in the list with the next at once. 
The last vector in the list should also be connected to the first. 
Calling connect_the_dots(plane, dino_vectors) should give the complete outline of the dinosaur, 
provided the dino_vectors are in the right order.
'''

'''
Note the use of mod(%) operator. Key point to remember is 1%10 == 1, 2%10 == 2 , n%m == n if n < m 
'''
def connect_the_dots(vectors):
    for x in range(0, len(vectors), 1):
        draw_segment(vectors[x] , vectors[(x+1)%len(vectors)])
    
```

```python
##Draw Dino 
fig, ax = plt.subplots()
draw_axes(ax)
draw_grid(ax)
draw_origin()
set_bounds(-8,8,-8,8)
draw_points(dino_vectors)
connect_the_dots(dino_vectors)
display()
```
